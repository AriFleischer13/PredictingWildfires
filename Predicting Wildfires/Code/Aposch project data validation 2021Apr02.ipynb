## Temporal binning of weather data.

We are unable to determine the timezone of the wildfires data. It may be in local time, which UTC-8 in California/Nevada for the winter months, or it may be in UTC. When a day starts in UTC, it will be 8 hours before that same day starts in California/Nevada. Our weather data is in UTC time. *If* our fires data is in local time, this is convenient for us, beacuse for a specific date, our weather features start 8 hours earlier than our fires. If our weather features are in local time, 

Nevertheless, we were unable to find which timezone the fires data are in, so in the worst-case scenario there would be on average 8 hours of misalignment on either side of each bin. The worst-case misalignment would be 16 hours per bin, which is too much for a bin of 24 hours. In order to be careful and reduce the percentage of time per bin that could be error, it would be **wise to bin up to longer time periods: try 72 hours (3 days), 5 days, and 10 days**.

Temporal binning also allows us to create new features, using the max, min, range, and standard deviation of each of our daily weather features. The 10-day binned dataset also mimics research by Cao et al in China, allowing us to compare our model to theirs.

Binned info should include start year, end year, start doy, end doy
For each base parameter: bin max, bin min, bin mean, bin range (=max-min), bin std



### Are we trying to mimic the Cao et al paper?

The Cao et al paper is invalid for real-time predictions, because they used weather features and wildfire ignitions from the same 10-day period, meaning they included weather measurements from just *after* an ignition as part of the input in order to "predict" it. For Cao et al, their methods were acceptable, as their ultimate goal was to create a wildfire susceptibility map for a province in China - they used the highest-probability predictions over 9 years from each location to assign that location a susceptibility.

If we have time, we can try for **two different goals**. One goal would be to see whether we can **match the predictions of Cao et al**, which means a susceptibility map for California/Nevada could be created that is as accurate as their map was for southern China. Another goal would be to see whether our model can **precidt wildfires *before* they happen** - i.e. based on to-the-moment weather data, predict whether a wildfire will happen in the next few days. This cannot be compared to what Cao et al did, and it requires that we join our fires data differently, but it may be a more compelling use-case for our model.

```python
data1 = merged_data.copy()

data1['longitude'] = data1['longitude'] - 360 # change longitude units to match fires dataset 
```

```python
# Create reference for various indexes
time_reference = merged_data.loc[:,['Year','DOY']].copy()
tr2 = time_reference.drop_duplicates().reset_index().reset_index()
tr2['index_3day'] = np.floor( tr2['level_0'] / 3 ).astype('int32')
tr2['index_5day'] = np.floor( tr2['level_0'] / 5 ).astype('int32')
tr2['index_10day'] = np.floor( tr2['level_0'] / 10 ).astype('int32')
tr2['macro_season'] = np.floor( tr2['DOY'] / 183 ).astype('int32')
tr2['month'] = np.floor( tr2['DOY'] / 30.5 ).astype('int32') + 1
tr2
```

```python
tr3 = tr2.loc[:,['Year','DOY','macro_season','month','index_3day','index_5day','index_10day']].copy()
```

```python
data4 = pd.merge(tr3, data1, how='inner', 
         left_on=['Year','DOY'], 
         right_on=['Year','DOY'])

data4

#Xy['fire'] = Xy['fire'].fillna(0)
```

```python
data_3day = data4.groupby(
    ['latitude','longitude','index_3day']).agg({'time':['first','last'],
                                                'DOY':['first','last'],
                                                'Year':['first','last'],
                                                'cwat':[np.min, np.max, np.mean, np.var],
                                                'r':[np.min, np.max, np.mean, np.var],
                                                'tozne':[np.min, np.max, np.mean, np.var],
                                                'gh':[np.min, np.max, np.mean, np.var],
                                                'pwat':[np.min, np.max, np.mean, np.var],
                                                'paramId_0':[np.min, np.max, np.mean, np.var],
                                                'pres':[np.min, np.max, np.mean, np.var, 'count'],
                                                'macro_season': lambda x: mode(x)[0],
                                                'month': lambda x: mode(x)[0]})
```
